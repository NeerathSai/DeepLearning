```python
from fast_ml.model_development import train_valid_test_split  
```

```python
X_train, y_train, X_valid, y_valid, X_test, y_test = train_valid_test_split(df, target = 'BodyFat', 
                                                                            train_size=0.7, valid_size=0.15, test_size=0.15)
```

## **MODEL1 with 1 HIDDEN LAYER WITH RELU ACTIVATION FUNCTION**

MODEL 1 with 1 Hidden Layer with L1 regularisation

---



```python
model1 = tf.keras.Sequential()
model1.add(tf.keras.layers.Flatten(input_shape=[14], name="inputLayer"))
model1.add(tf.keras.layers.Dense(128, activation="relu", kernel_regularizer = tf.keras.regularizers.l1(), name="HiddenLayer1"))
model1.add(tf.keras.layers.Dense(1, activation="linear", name="outputLayer"))
model1.summary()
```

```python
for epochs in (10,20,30,40):
  for lr in (1, 0.3, 0.1, 0.01,0.03,0.001,0.0001,0.00001):
    print("For number of epochs: ",epochs )
    print("For learning rate: ",lr)

    model1.compile(loss="mean_squared_error",optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
                   metrics=[tf.keras.metrics.RootMeanSquaredError()])
    batch_size = 32
    history = model1.fit(X_train, y_train,batch_size=batch_size, epochs=epochs, validation_data=(X_valid,y_valid))
    model1.evaluate(X_test,y_test)
```
